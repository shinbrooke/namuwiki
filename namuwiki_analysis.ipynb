```python
#대중문화 및 서브컬처 모두 하나의 데이터프레임으로
culture_names = ['readymadelife', 'angrybird', 'ssglanders', 'myname', 'gameserver', 'itzy', 'crashlandingonyou', 'maplephantom', 'skycastle', 'transformer']
df_final = pd.DataFrame()

for i in range(len(culture_names)):
  #readfile
  filename = '/content/drive/My Drive/2022-2 데이터 저널리즘 팀플/대중문화 및 서브컬처/namu_c_'+str(culture_names[i])+'.txt'
  df =pd.read_csv(filename, header=None, sep='\t')
  #split file
  df.columns = ['col1']
  df0 = df.col1.str.split(' ', expand=True)
  #add column names
  c = list(range(len(df0.columns)))
  cn = []
  for a in c:
    c2 = "c"+str(c[a])
    cn.append(str(c2))
  df0.columns = cn
  #create comment column
  for c in cn:
    df0[c] = df0[c].fillna("")
  df0['other'] = df0.iloc[:,17:len(cn)].apply(lambda x: ''.join(x), axis=1)
  #extract columns and name them
  df1 = df0[['c0','c1','c14','c15','c16','other']]
  df1.columns = ['date','time','code','change','uname','other']
  df1['page'] = str(culture_names[i])
  df1['category'] = 'culture'
  #create total dataframe
  df_final = pd.concat([df_final,df1], ignore_index=True)
```

```python
df_final.to_csv('/content/drive/MyDrive/2022-2 데이터 저널리즘 팀플/대중문화 및 서브컬처/dataculture.csv')
df_final
```

```python
#학문 모두 하나의 데이터프레임으로
academic_names = ['spotlight', 'aesthetic', 'call', 'epidemic', 'greekromanmyth', 'hungarianrevolution', 'montyhall', 'imjin', 'officiallanguage', 'pascaltriangle']
df_final = pd.DataFrame()

for i in range(len(academic_names)):
  #readfile
  filename = '/content/drive/My Drive/2022-2 데이터 저널리즘 팀플/학문/namu_a_'+str(academic_names[i])+'.txt'
  df =pd.read_csv(filename, header=None, sep='\t')
  #split file
  df.columns = ['col1']
  df0 = df.col1.str.split(' ', expand=True)
  #add column names
  c = list(range(len(df0.columns)))
  cn = []
  for a in c:
    c2 = "c"+str(c[a])
    cn.append(str(c2))
  df0.columns = cn
  #create comment column
  for c in cn:
    df0[c] = df0[c].fillna("")
  df0['other'] = df0.iloc[:,17:len(cn)].apply(lambda x: ''.join(x), axis=1)
  #extract columns and name them
  df1 = df0[['c0','c1','c14','c15','c16','other']]
  df1.columns = ['date','time','code','change','uname','other']
  df1['page'] = str(academic_names[i])
  df1['category'] = 'academic'
  #create total dataframe
  df_final = pd.concat([df_final,df1], ignore_index=True)

df_final.to_csv('/content/drive/MyDrive/2022-2 데이터 저널리즘 팀플/학문/dataacademic.csv')
df_final
```
